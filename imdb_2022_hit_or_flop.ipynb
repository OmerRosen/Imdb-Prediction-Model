```python
import requests
from IPython.core.display import HTML
import json
from time import sleep
import re
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
import time
```

```python
base_url = "https://www.imdb.com"
movie_search_url = f"{base_url}/search/title/?title_type=feature&release_date=2022-01-01,2022-12-31&num_votes=20,&countries=us&languages=en&start=1&ref_=adv_nxt"
```

```python
list_of_50_movies = requests.get(f'{movie_search_url}')
```

```python
list_of_50_movies_soup = BeautifulSoup(list_of_50_movies.text, 'html.parser').find_all('div',{'class':'lister-item mode-advanced'})
```

```python
movie = list_of_50_movies_soup[1]
def get_basic_details(movie):
  basic_details_dict = {}
  basic_details_dict['img_link'] = movie.find('img',{'class':'loadlate'})['src']
  movie_path = movie.find('a')['href']
  basic_details_dict['movie_url'] = f"{base_url}{movie_path}"
  basic_details_dict['movie_id'] = movie_path.split('/')[-2]
  basic_details_dict['movie_title'] = movie.find('img',{'class':'loadlate'})['alt']
  movie_p = movie.find('p',{'class':'text-muted'})
  basic_details_dict['age_certificate'] = movie_p.find('span',{'class':'certificate'}).text
  basic_details_dict['runtime_min'] = movie_p.find('span',{'class':'runtime'}).text.split(' ')[0]
  basic_details_dict['genres'] = movie_p.find('span',{'class':'genre'}).text.strip()
  basic_details_dict['rating_count'] = movie.find('meta',{'itemprop':'ratingCount'})['content']
  basic_details_dict['rating_imdb'] = movie.find('meta',{'itemprop':'ratingValue'})['content']
  basic_details_dict['movie_desc'] = movie.find_all('p',{'class':'text-muted'})[-1].text.strip()
  basic_details_dict['metascore'] = movie.find('span',{'class':'metascore favorable'}).text.strip()
```
